#**LISTAS EN PYTHON**

Enunciado: Crea una lista llamada invitados con los nombres: "Ana", "Luis", "Pedro". Luego, añade a "María" al final de la lista y cambia a "Luis" por "Juan".

**2. Suma de una lista**

Enunciado: Tienes la lista de precios [10, 5, 20]. Usa la función sum() de Python para mostrar el total de la cuenta.

**3. ¿Cuántos hay?**

Enunciado: Crea una lista con los colores: "rojo", "azul", "rojo", "verde", "rojo". Usa un método de listas para contar cuántas veces aparece el color "rojo".

**4. El primero y el último**

Enunciado: Dada la lista frutas = ["manzana", "pera", "platano", "kiwi"], imprime solo la primera y la última fruta (usa índices).

**5. Ordenando la fila**
